```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
location = r"https://raw.githubusercontent.com/Walker2921/Data-Analytics/master/Student's%20Performance%20Project/StudentsPerformance%20(4)%20(1).csv"
raw_data_set = pd.read_csv(location)
raw_data_set.shape
```

```python
raw_data_set.describe(include = "all")
```

```python
def info_table(raw_data_set):
    unique_values = []
    not_null_count = []
    null_count = []
    is_null = []
    for i in list(raw_data_set.columns.values):
        unique_values.append(raw_data_set[i].value_counts().count())
        null_count.append(raw_data_set[i].isnull().sum())
        k = raw_data_set.shape[0] - null_count[-1]
        not_null_count.append(str(k) + " non-null")
        is_null.append(str(round(raw_data_set[i].isnull().sum()/len(raw_data_set)*100, 2))+"%")
    
    temp = {"unique_values" : unique_values, 
            "not_null_count" : not_null_count, 
            "is_null" : is_null, 
            "null_count" : null_count, 
            "Dtypes" : raw_data_set.dtypes}
    df = pd.DataFrame(temp, columns = ["unique_values", 
                                       "not_null_count", 
                                       "is_null", 
                                       "null_count", 
                                       "Dtypes"])
    return df
```

```python
info_table(raw_data_set)
```

### The above table concludes that there aren't any missing values in the Dataset, so we won't be needed to perform Data Manipulation on the Dataset.

```python
data_set = raw_data_set
```

# Data Analysis 

```python
column_names = list(data_set.columns)
column_names.pop(), column_names.pop(), column_names.pop()
column_values = []
for i in range(0, 5):
    temp = data_set.iloc[:, i].unique()
    column_values.append(list(temp))
k = ["math score", 'reading score', 'writing score']
```

> ## 1. Gender

```python
plt.figure(figsize = (15, 5))
val = column_names[0]
for i in enumerate(k):
    plt.subplot(1, 3, i[0]+1)
    for j in column_values[column_names.index(val)]:
        sns.distplot(data_set[data_set[val]==j][i[1]], 
                    hist = False, label = j)
```

### The Males performing well in the case of Mathematical skills, whereas Females takeover when it comes to Reading and Writing skills.

> ## 2. Race/Ethnicity

```python
plt.figure(figsize = (15, 5))
val = column_names[1]
for i in enumerate(k):
    plt.subplot(1, 3, i[0]+1)
    for j in column_values[column_names.index(val)]:
        sns.distplot(data_set[data_set[val]==j][i[1]], 
                    hist = False, label = j)
```

```python
plt.figure(figsize = (20, 4))
val = column_names[1]
p = 0
for i in ['group B', 'group C', 'group A', 'group D', 'group E']:
    plt.subplot(1, 5, p+1)
    for j in enumerate(k):
        sns.distplot(data_set[data_set[val]==i][j[1]], 
                    hist = False, label = j[1])
        plt.xlabel(i)
    p+=1
```

### It is clear from above that overall performance of Group A is quite dissatisfying followed by Group B, whereras in the case of good performance it looks like Group E and Group D are almost at a tie, and after all Group C is performing average for all cases.

> ## 3. Parental level of Education

```python
plt.figure(figsize = (15, 5))
val = column_names[2]
for i in enumerate(k):
    plt.subplot(1, 3, i[0]+1)
    for j in column_values[column_names.index(val)]:
        sns.distplot(data_set[data_set[val]==j][i[1]], 
                    hist = False, label = j)
```

```python
plt.figure(figsize = (20, 7))
val = column_names[2]
p = 0
for i in ["bachelor's degree", 'some college', "master's degree", "associate's degree", 'high school', 'some high school']:
    plt.subplot(2, 3, p+1)
    for j in enumerate(k):
        sns.distplot(data_set[data_set[val]==i][j[1]], 
                    hist = False, label = j[1])
        plt.xlabel(i)
    p+=1
```

### It looks like students whose parents have Master's Degree are overall performing the best, followed by Bachelor's Degree. It seems that children with parents having Associate's Degree are not performing upto the mark in the case of Mathematical skills whereas they perform quite well when it comes to Reading and Writing Skills, followed by Some College and High School, and as expected children with parents with Some High Schools  educational background are performing the Worst.
